```python
import MeCab

text=[]
with open("../data/neko.txt", "r") as f:
    for line in f:
        if line.strip()!="":
            text.append(line.strip())

keitaiso = MeCab.Tagger('-Ochasen')    #分かち書き
#neo_keitaiso = MeCab.Tagger('-Ochasen -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd') #追加辞書を適用

with open("../data/neko.txt.mecab", "w") as f:
    for line in text:
        f.write(keitaiso.parse(line))
```

```python
#30
result=[]
with open("../data/neko.txt.mecab", "r") as f:
    for line in f:
        result.append(line)

def extract_sentence(text):
    result = []
    dic={}
    sentence=[]
    for line in text:
       
        if line=="EOS\n":
            result.append(sentence)
            sentence=[]
            
            continue
        dic["surface"]=line.split("\t")[0]
        dic["base"]=line.split("\t")[2]
        dic["pos"]=line.split("\t")[3]
        dic["pos1"]=line.split("\t")[4]
        sentence.append(dic)
        dic={}
    return result

extract_sentence(result[0:100])[1]
```

```python
#31
all_result=extract_sentence(result)
ans=[]
for sentence in all_result:
    for word in sentence:
        if "動詞-" in word["pos"]:
            ans.append(word["surface"])
print(ans)
```

```python
#34
def extract_none(text):
    nones=[]
    flag=False
    for sentence in text:   
        none=""
        for word in sentence:
            if "名詞" in word["pos"]:
                none+=word["surface"]
                flag=True
            else:
                if flag:
                    nones.append(none)
                    none=""
                    flag=False
    return nones
extract_none(all_result)   
```

```python
#35
import collections
ans=[]
for sentence in all_result:
    for word in sentence:
        ans.append(word["base"])
sorted(collections.Counter(ans).items(), key=lambda x: -x[1])
```

```python
#36
import matplotlib.pyplot as plt
words=[]
times=[]
for x, y in sorted(collections.Counter(ans).items(), key=lambda x: -x[1])[:10]:
    words.append(x)
    times.append(y)
plt.bar(words, times)
plt.show()
```
